```python
image_batch, digits_batch, letters_batch = Donkey.build_batch('data/train.tfrecords', num_examples=100, batch_size=36, shuffled=True)
```

```python
sess = tf.InteractiveSession()
coord = tf.train.Coordinator()
threads = tf.train.start_queue_runners(coord=coord)
```

```python
image_batch_val, digits_batch_val, letters_batch_val = sess.run([image_batch, digits_batch, letters_batch])
image_batch_val = (image_batch_val / 2.0) + 0.5
```
